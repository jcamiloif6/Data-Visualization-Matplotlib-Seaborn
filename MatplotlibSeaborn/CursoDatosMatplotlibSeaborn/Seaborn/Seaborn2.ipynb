```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", context="talk", palette="tab10")

intakes = sns.load_dataset("tips")

intakes
```

# <font color=#99FFFF>Graficas Categoricas de Dispersion</font>

## <font color='#FF9933'>Grafica stripplot</font>

```python
sns.catplot(x="day", y="total_bill", data=intakes,
           kind="strip", hue="size", col="smoker",
           s=7)
plt.show()
```

## <font color='#FF9933'>Grafica swarmplot</font>

```python
sns.catplot(x="day", y="total_bill", data=intakes,
           kind="swarm", hue="size", col="smoker",
           s=7)
plt.show()
```

# <font color=#99FFFF>Graficas Categoricas de Distribucion</font>

## <font color='#FF9933'>Grafica boxplot</font>

```python
sns.catplot(x="day", y="total_bill", data=intakes,
           kind="box", col="smoker")

plt.show()
```

## <font color='#FF9933'>Grafica boxenplot</font>

```python
sns.catplot(x="day", y="total_bill", data=intakes,
           kind="boxen", col="smoker")

plt.show()
```

## <font color='#FF9933'>Grafica violinplot</font>

```python
sns.catplot(x="day", y="total_bill", data=intakes,
           kind="violin")

sns.catplot(x="day", y="total_bill", data=intakes,
           kind="violin", hue="smoker")

sns.catplot(x="day", y="total_bill", data=intakes,
           kind="violin", hue="smoker", split=True)

plt.show()
```

# <font color=#99FFFF>Graficas Categoricas de Estimacion</font>

## <font color='#FF9933'>Grafica pointnplot</font>

```python
sns.catplot(x="day", y="total_bill", data=intakes,
           kind="point", col="smoker")

plt.show()
```

## <font color='#FF9933'>Grafica barplot</font>

```python
sns.catplot(x="day", y="total_bill", data=intakes,
           kind="bar", hue="size")

plt.show()
```

## <font color='#FF9933'>Grafica countplot</font>

```python
sns.catplot(x="day", data=intakes,
           kind="count", hue="size")

plt.show()
```
