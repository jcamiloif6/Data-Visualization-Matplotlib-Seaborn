```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.set_theme(style="darkgrid", context="notebook")

relation = sns.relplot(data=penguins, height=7, aspect=1.2,
                       markers=["s",".","p","d"])
relation.set(title="Analisis de Pinguino")
relation.set_ylabels("Variables")
relation.set_xlabels("Numero de Pinguinos")
#plt.grid(False)
plt.show()
```

```python
print(">>>>>>>>>>>>>>> SIMPLE")

sns.relplot(data=penguins, x="bill_length_mm", y="flipper_length_mm")
plt.show()

print(">>>>>>>>>>>>>>> MATIZ Y TAMAÃ‘O")

sns.relplot(data=penguins, x="bill_length_mm", y="flipper_length_mm",
            hue="body_mass_g", s=150)
plt.show()

sns.relplot(data=penguins, x="bill_length_mm", y="flipper_length_mm",
            hue="species", s=150)
plt.show()

print(">>>>>>>>>>>>>>> REJILLA, TAMAÃ‘O PERSONALIZADO Y ESTILO")

sns.relplot(data=penguins,
            x="bill_length_mm",
            y='flipper_length_mm',
            hue="species",
            row="sex",
            size="body_mass_g",
            sizes=(100,500),
            style="sex")
plt.show()

sns.relplot(data=penguins,
            x="bill_length_mm",
            y='flipper_length_mm',
            hue="species",
            col="island",
            row="sex",
            size="body_mass_g",
            sizes=(100,500),
            style="sex")
plt.show()
```

**GRAFICO DE LINEAS**

```python
sns.relplot(data=penguins,
            x="species",
            y="body_mass_g",
            kind="line",
            hue="sex",
            style="sex",
            ci="sd")

plt.show()

sns.relplot(data=penguins,
            x="bill_depth_mm",
            y="flipper_length_mm",
            kind="line",
            hue="species",
            style="species",
            ci=99)

plt.show()
```

## **COMBINANDO GRAFICOS**

**Lineas y Dispersion**

```python
figura, ejes = plt.subplots(1,2,figsize=(15,5))

sns.lineplot(data=penguins,
            x="species",
            y="body_mass_g",
            hue="sex",
            style="sex",
            errorbar='sd',
            ax=ejes[0])

sns.scatterplot(data=penguins, x="bill_length_mm", y="flipper_length_mm",
            hue="species", s=150, ax=ejes[1])

plt.show()
```
