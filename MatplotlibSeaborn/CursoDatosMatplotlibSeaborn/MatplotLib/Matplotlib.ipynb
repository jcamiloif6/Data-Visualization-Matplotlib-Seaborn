```python
import numpy as np

years = [2011,2012,2013,2014,2015,
         2016,2017,2018,2019,2020]

nivel1 = np.random.rand(10)*100
nivel2 = np.random.rand(10)*200 + 100
nivel3 = np.random.rand(10)*300 + 200
nivel4 = np.random.rand(10)*400 + 300
nivel5 = np.random.rand(10)*500 + 400
```

```python
# Series de datos con color, marcador, etiqueta y estilo

plt.figure(figsize=(12,8))

plt.plot(years, nivel1, label="Nivel 1", color='purple', marker="<", linestyle="-")
plt.plot(years, nivel2, label="Nivel 2", color='red', marker="*", linestyle="--")
plt.plot(years, nivel3, label="Nivel 3", color='blue', marker="^", linestyle="-.")
plt.plot(years, nivel4, label="Nivel 4", color='black', marker=".", linestyle=":")
plt.plot(years, nivel5, label="Nivel 5", color='green', marker="+", linestyle=" ")

# Activar Leyendas
plt.legend()

# Titulo de grafica
plt.title("Examen de Certificacion")

# Etiqueta de los ejes
plt.xlabel("Año de realizacion del examen")
plt.ylabel("Puntaje")

# Eje de Puntaje de 0 a 1000 en incrementos de 50 (ticks mayores)
plt.yticks(np.arange(0,1001, 50))

# Activar Cuadricula
plt.grid()

# Activar marcas(ticks) menores
plt.minorticks_on()

plt.show()
```

**GRAFICA BARRAS** **HORIZONTALES**

```python
eje_x = np.arange(0,10)

plt.bar(eje_x, nivel5, width=1/5)
plt.bar(eje_x+0.2, nivel4, width=1/5)
plt.bar(eje_x+0.4, nivel3, width=1/5)
plt.bar(eje_x+0.6, nivel2, width=1/5)
plt.bar(eje_x+0.8, nivel1, width=1/5)

plt.show()
```

**GRAFICA DE BARRAS HORIZONTALES**

```python
plt.barh(eje_x+0.8, nivel5, height = 1/5)
plt.barh(eje_x+0.6, nivel4, height = 1/5)
plt.barh(eje_x+0.4, nivel3, height = 1/5)
plt.barh(eje_x+0.2, nivel2, height = 1/5)
plt.barh(eje_x, nivel1, height = 1/5)

plt.show()
```

**GRAFICA DE BARRAS VERTICALES APILADAS**

```python
plt.bar(eje_x,nivel5)
plt.bar(eje_x,nivel4, bottom=nivel5)
plt.bar(eje_x,nivel3, bottom=nivel4+nivel5)
plt.bar(eje_x,nivel2, bottom=nivel3+nivel4+nivel5)
plt.bar(eje_x,nivel1, bottom=nivel2+nivel3+nivel4+nivel5)

plt.show()
```

**GRAFICO DE DISPERSION**

```python
plt.scatter(eje_x, nivel1, marker='*')
plt.scatter(eje_x, nivel2)
plt.scatter(eje_x, nivel3)
plt.scatter(eje_x, nivel4)
plt.scatter(eje_x, nivel5)

plt.show()
```

**GRAFICO DE PIE**

```python

plt.pie(nivel1, labels=["Mexico","Colombia","España","Estados Unidos","Ecuador",
          "Argentina","Venezuela","Puerto Rico", "Honduras","Uruguay"])

plt.show()
```

**Combinar Graficas**

```python
plt.bar(eje_x, nivel1, width=1/5)
plt.bar(eje_x+0.2, nivel1, width=1/5)

plt.plot(years, nivel3)

plt.scatter(eje_x, nivel4)
plt.scatter(eje_x, nivel5)

plt.show()
```
